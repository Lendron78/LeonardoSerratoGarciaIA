```python
import numpy as np
import cv2 as cv

# Carga la Imagen
image = cv.imread(r'D:\Escuela Betin\9no Sem\3. Inteligencia Artificial\Wally\Pruebas\Wally2.jpg')

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Carga el xml
wally = cv.CascadeClassifier(r'D:\Escuela Betin\9no Sem\3. Inteligencia Artificial\Wally\classifier\cascade.xml')

# Parametros para detectar a wally
wally_detections = wally.detectMultiScale(gray, scaleFactor=1.08, minNeighbors=3, minSize=(3,4))

# Dibuja Rectangulos y pone el texto
for (x, y, w, h) in wally_detections:
    cv.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv.putText(image, 'Wally', (x, y - 10), cv.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Muestra el resultado
cv.imshow('Donde esta Wally?', image)
cv.waitKey(0)
cv.destroyAllWindows()
```
